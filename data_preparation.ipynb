```python
def aggregate_csv_data(target_folder, save=True, filetype='csv'):
    all_data = pd.DataFrame()
    files_to_aggregate = ['Friday-16-02-2018_TrafficForML_CICFlowMeter.csv.parquet', 'Thursday-15-02-2018_TrafficForML_CICFlowMeter.csv.parquet']
    for file in files_to_aggregate:
        file_path = f'{target_folder}/{file}.{filetype}'
        print(file_path)
        df = pd.DataFrame()
        if filetype == 'csv':
            df = pd.read_csv(file_path)
        print(df.shape)
        print(f'{df["Label"].value_counts()}\n')
        all_data = pd.concat([df, all_data], ignore_index=True)
    print('ALL DATA')
    duplicates = all_data[all_data.duplicated(subset=all_data.columns.difference(['Label', 'Timestamp']))]
    print('Removed duplicates after aggregating:')
    print(duplicates.Label.value_counts())
    print('Resulting Dataset')
    all_data.drop(duplicates.index, axis=0, inplace=True)
    all_data.reset_index(inplace=True, drop=True)
    print(all_data.shape)
    print(f'{all_data["Label"].value_counts()}\n')
    
    if save:
        malicious = all_data[all_data.Label != 'Benign'].reset_index(drop=True)
        benign = all_data[all_data.Label == 'Benign'].reset_index(drop=True)
        all_data.to_csv(f'agg_csv/all_data.csv', index=False)
        malicious.to_csv(f'agg_csv/all_malicious.csv', index=False)
        benign.to_csv(f'agg_csv/all_benign.csv', index=False)
```

```python
aggregate_csv_data('csved', save=True, filetype='csv')
```
